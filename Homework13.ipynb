# Transposed convolution in theory
**Transposed convolution** is a process that generates the output feature map, which has greater spatial dimensions than the input feature map. It is widely used e.g. in a decoder phase in the encoder-decoder architectures or in GAN's (Generative Adversarial Networks) to turn small vectors into larger images.

It is an opposite process to classic convolution, where we reduce the dimension of the data to extract some characteristic features. In transposed convolution we take the reduced feature map and try to recover the original image from it.

# How does it work?
Assume that we have an input two-dimensional tensor of dimensions $t_x \times t_y$ and a kernel of dimensions $k_x \times k_y$.

At each step, we select one coefficient from the output tensor and treat it as a scalar. We multiply the entire kernel by this scalar. Next, we move one position to the right (or start a new row when necessary) and repeat the operation until all coefficients have been processed. In this way, we obtain $t_x t_y$ copies of the kernel, denoted by $K_{ij}$, each multiplied by a scalar.

We then consider a zero matrix of dimensions $(t_x + k_x) \times (t_y + k_y)$. Into this matrix, we add the matrices $K_{ij}$ one by one. The first matrix is placed in the upper-left corner. In each subsequent step, the placement is shifted by one position to the right (or starts at the beginning of the next row when the end of the current row is reached).

# How do we make it even more interesting?
Just as in case of regular convolution we can control the spatial dimensions of the output map by parameters like **kernel size**, **padding** or **stride**. Although we have to be careful, because this time this terms often does the opposite action, as in the regular case. Let's look at the differences:

- Kernel size -- size of the kernel matrix. As we have seen above, the output map has dimensions (t_x + k_x) \times (t_y + k_y)$, so the larger the kernel, the larger the output map is.

- Padding -- in contrast to regular convolution, it removes first and last rows and columns. When we set padding $= n$, we repeat this process $n$ times. For example padding $= 1$ will turn $5 \times 5$ matrix into $3 \times 3$ matrix, whereas padding $= 2$ will turn $5 \times 5$ matrix into $1 \times 1$ matrix.

- Stride -- again it has the opposite action as in the regular convolution. In *transposed convolution* it determines the spacing between consecutive kernels K_{ij} in the output matrix. For example, when stride $= 2$, we move 2 steps to the right, instead of 1.
