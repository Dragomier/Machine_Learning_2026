We will show that first and third formulation are equivalent, but they are not equivalent to second one.

**Equivalence of first and third**

Suppose that we have partially built tree $T$ and let $(R_p, X_j, t) = \text{argmin}_{\text{Parent}, X_j, t}\{\sum_{m=1}^{2} \sum_{i : x_i \in R_m(X_j, t)} (y_i - \bar{y}_{R_m})^2\}$. Then sum of residuals in other regions remains the same, so
$$
\text{Total RSS sum} = \sum_{m \in \text{Regions}} \sum_{i \in R_m} \Bigl(y_i - \overline{y}_{R_m}\Bigr)^2 = \underbrace{\sum_{m \in \text{Regions} \setminus \{R_p\}} \sum_{i \in R_m} (y_i - \overline{y}_{R_m})^2}_{\text{fixed}} + \sum_{k=1}^{2} \sum_{i : x_i \in R_k(X_j, t)} (y_i - \overline{y}_{R_k})^2
$$
We concluse that if our triple minimizes the second part of the sum, then because the rest is fixed, the triple minimizes also total RSS sum.

**Non-equivalence of first and second**

Suppose that we have two regions, one with $RSS = 5$ and second one with $RSS = 200$. Then the best split in the first partition may give regions $R_{11}$ with $RSS = 1.5$ and $R_12$ with $RSS = 2$ and he best split in the second partition may give regions $R_{21}$ with $RSS = 90$ and $R_22$ with $RSS = 80$. Then sum of children's RSS is smaller in region one, so local RSS will choose to split in the first region, whereas RSS Gain will chose to partition second region, because in the second region there is bigger change.
