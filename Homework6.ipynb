We will show that Local RSS Minimization and RSS Gain Maximization are equivalent, but they are not eqiuvalent to Total RSS Minimization.

Firstly, fix a parent region $R_p$. Let:

$$
X = \sum_{i : x_i \in R_p}(y_i - \bar{y}_{R_p})^2 - \sum_{m=1}^{2} \sum_{i : x_i \in R_m(X_j, t)} (y_i - \bar{y}_{R_m})^2 \text{ -- expression that RSS Gain maximizes}
$$

$$
Y = \sum_{m=1}^{2} \sum_{i : x_i \in R_m(X_j, t)} (y_i - \bar{y}_{R_m})^2 \text{ -- expression that local RSS minimizes }
$$

Observe that
$$
X + Y = \sum_{i : x_i \in R_p}(y_i - \bar{y}_{R_p})^2  \text{  -- constant value for fixed region } R_p
$$
From that we can conclude that if we choose a split $(X_j, t)$ that maximizes X, then it will also minimize Y and vice versa. Because the equivalence holds for each parent, it also holds when optimizing over all parents.
