```python
import torch
import model_from_lab
```

```python
neural_net = model_from_lab.MLP()
neural_net.load_state_dict(torch.load("model.pth"))
```

```python
import matplotlib.pyplot as plt
random_noise_photos = torch.randn(size = [10, 1, 28, 28], requires_grad = True)
targets = torch.arange(10)

for param in neural_net.parameters():
    param.requires_grad = False

optimizer = torch.optim.Adam([random_noise_photos], lr=0.01)
epochs_state = []
num_epochs = 0
for epochs in range(50):
    optimizer.zero_grad()
    neural_output = neural_net(random_noise_photos)
    loss = torch.nn.functional.cross_entropy(neural_output, targets, reduction = "mean")
    loss.backward()
    optimizer.step()
    epochs_state.append(random_noise_photos.clone().detach())
    score = 0
    predictions = torch.argmax(neural_output, dim=1)
    if torch.all(predictions == targets):
        print(f"Network learns its numbers after {epochs} epochs")
        break
```

```python
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
%matplotlib notebook

def make_animation(number, num_epochs, epoch_state):
    fig, ax = plt.subplots(1, 1, figsize=(5, 5))
    plot = ax.imshow(epoch_state[0][number].detach().cpu().squeeze(), cmap = "gray")

    def update(frame):
        plot.set_data(epoch_state[frame][number].detach().cpu().squeeze())
        return [plot]
    anim = FuncAnimation(fig, update, frames=num_epochs, interval=100, blit=True)
    display(HTML(anim.to_jshtml()))
```

```python
make_animation(0, len(epochs_state), epochs_state)
```
